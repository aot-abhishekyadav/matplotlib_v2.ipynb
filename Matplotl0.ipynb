```python
'''
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#it is mostly use in jupyter notebook
```

```python
# if we use %matplotlib inline then we don't have to write plt.show and in collab it is not necessary to write plt.show
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# plotting using matplotlib
# plt scatter
plt.scatter(x,y, c = 'g', )  # for scatter x and y must be the same size
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
plt.show()
#plt.savefig('2d.png')
```

```python
# plotting using matplotlib
# plt scatter
plt.scatter(x,y, c = 'r', s = 100)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
plt.show()
#plt.savefig('2d.png')
```

```python
# using plt plot
y = x*x
plt.plot(x, y, 'r*', linestyle = '--',)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
plt.show()
```

```python
# using plt plot
y = x*x
plt.plot(x, y, 'r*', linestyle = '--', linewidth = 2, markersize = 10)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Graph')
plt.show()
```

```python
'''
Differnce between scatter and plot
Scatter plots are used to display individual data points, showing the relationship between two variables. plots are used to connect data points with lines and scatter are used to show data through the help of points
'''
```

```python
# creating subplot
plt.subplot(2,2,1) # 2 - row, 2 - column, 1 - poistion
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*', linewidth = 2, markersize = 5)
plt.subplot(2,2,3)
plt.plot(x,y,'bo') # making scatter with the help of plot
plt.subplot(2,2,4)
plt.plot(x,y,'yo', linewidth = 3, markersize = 7)
plt.show()
```

```python
x = np.arange(1,11)
y = 4 * x + 4
plt.plot(x,y,'r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Equation graph')
plt.show()
```

```python
# making sine graph
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Sine Wave')
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y, 'g^')
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('Sine Wave')
plt.show()
```

```python
# making sine and cosine wave form
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
# making 1st subplot
plt.subplot(2,1,1) # 2-row, 1 - column, 1 - position
plt.plot(x,y_sin)
plt.title('Sine wave')
# making 2nd subplot
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine wave')
plt.show()
```

```python
# Bar plot
x = [2,6,8]
y = [4,12,16]

x1 = [3,5,9]
y1 = [9,10,18]

plt.bar(x,y, color='b')
plt.bar(x1,y1, color='g')

plt.title('Bar Graph')
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.show()
```

```python
#histogram
a = np.array([20,80,5,40,50,70,55,54,11,15,51,5,79,31,25])
plt.hist(a)
plt.title("histogram")
plt.show()
```

```python
#histogram
a = np.array([20,80,5,40,50,70,55,54,11,15,51,5,79,31,25])
plt.hist(a,bins = 15) # bins is using make gap between on x axis
plt.title("histogram")
plt.show()
```

```python
# piechart
labels = ['PYTHON', 'C++', 'JAVA', 'HTMl']
sizes = [250,150,100,50]
colors = ['gold','yellow','blue', 'green']

plt.pie(sizes, labels = labels, colors = colors, autopct= '%1.1f%%', shadow = False)
plt.axis('equal')
plt.show()
```

```python
# piechart
labels = ['PYTHON', 'C++', 'JAVA', 'HTMl']
sizes = [250,150,100,50]
colors = ['gold','yellow','blue', 'green']
explode = (0.4,0.1,0,0) # explode is use to make gap between chart
plt.pie(sizes, explode = explode, labels = labels, colors = colors, autopct= '%1.2f%%', shadow = True)
plt.axis('equal')
plt.show()
```

```python
# Boxplot
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.boxplot(data, vert = True, patch_artist = True)
plt.show()
```
